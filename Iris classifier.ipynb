```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
sns.set_style("white")
sns.kdeplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue="Species", fill=True)
```

```python
sns.violinplot(data=iris, x='Species', y='PetalLengthCm')
plt.title('Violin Plot of Petal Length by Species')
plt.show()
```

```python
sns.violinplot(data=iris, x='Species', y='SepalLengthCm')
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

```python
sns.swarmplot(data=iris, x='Species', y='PetalWidthCm')
plt.title('Swarm Plot of Petal Width by Species')
plt.show()
```

```python
sns.swarmplot(data=iris, x='Species', y='SepalWidthCm')
plt.title('Swarm Plot of Sepal Width by Species')
plt.show()
```

```python
sns.boxplot(data=iris, x='Species', y='PetalLengthCm')
plt.title('Box Plot of Petal Length by Species')
plt.show()
```

```python
sns.boxplot(data=iris, x='Species', y='SepalLengthCm')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
iris['Species_encoded'] = encoder.fit_transform(iris['Species'])

plt.figure(figsize=(8, 6))
sns.heatmap(iris.drop('Species', axis=1).corr(), annot=True, cmap='magma', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
